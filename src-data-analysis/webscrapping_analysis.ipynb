**How it works:**

1st - Run script on Pyhton on a number of different days to gather the data for the same products, creating a historical series.

2nd - Read all tabs from the excel file previously created, creating a dataframe for all products and all it´s data collected
(most importantly their prices)

3rd - Create new dataframe to show only once each product, so we can put all the prices collected for each product on the
same line in order to create a historical price line to see how the price for each product have changed.

```python
#First Part - Read, prepare and clean data.
```

```python
import pandas as pd
```

```python
#read all tabs
prices = pd.read_excel('CRC_prices.xlsx', sheet_name=None, index_col=0)
prices
```

```python
#concatenate all tabs under one dataframe
all_days_prices = pd.concat(prices.values())
all_days_prices.head()
```

```python
#Display the full dataframe
#pd.set_option('display.max_rows', None)
```

```python
#Set the name of products as index to be able to use .loc later
#all_days_prices.set_index('PRODUCT', inplace=True)
```

```python
#Create new df 'uniques' to show only one time all products 
uniques = pd.DataFrame(index=pd.unique(all_days_prices['PRODUCT']))
uniques
```

```python
len(uniques)
```

```python
#Adding a column 'count' to show how many price points we have for each product
uniques['count'] = all_days_prices['PRODUCT'].value_counts()
uniques
```

```python
#Create new df 'uniques_2' to preserve 'uniques' untouched
uniques_2 = uniques
uniques_2
```

```python
#Creating one empty column for each price point (historical serie) for each product.
for prod in uniques_2.index:
    for count in range(len(all_days_prices.loc[all_days_prices['PRODUCT'] == prod])):
        uniques_2['price ' + str(count)] = ''
```

```python
uniques_2
```

```python
#NOTE: WE HAVE GATHERED DATA FOR 44 DAYS (NOT CONSECUTIVE). 
#NOTE 2: SOME PRODUCTS DIDN´T HAVE ANY INFORMATION ON SOME OF THESE 44 OCCASIONS, PROBABLY BECAUSE THEY WERE OUT STOCK.
#THAT´S WHY WE´LL HAVE MANY BLANK DAYS ON A NUMBER OF PRODUCTS.
#NOTE 3: ALSO, IT MIGHT HAPPEN THAT ON THE BEGGINING OF THE SERIES WE HAD THE PRICE FOR A PRODUCT, THEN IT WAS OUT OF STOCK
#FOR SOME DAYS, THEN IT WAS BACK ON. WHEN THIS HAPPENS ALL THE BLANK SPOTS IN THE MIDDLE WILL BE FILLED WITH THE LAST PRICES
#WHEN THE PRODUCT WAS AVAILABLE AGAIN. THIS WOULD AFFECT THE THE MATCHING OF THE PRICE AT AN SPECIFIC DATE (IF WE INTEND TO
#MATCH DATE AND PRICE IN THE FUTURE).

#Filling up the previously created columns with the corresponding prices. 
#Cycle through every product in uniques_2
for prod in uniques_2.index:
    #For each of these products we´ll fill up the price columns using the prices collected on the original scraping (CRC_prices)
    for count in range(len(all_days_prices['PRODUCT'])):        
        try:
            #Each product will receive it´s prices on a different price column
            
            #Select product and price column on 'uniques_2' and fill that cel using it´s corresponding price on the original df
            uniques_2.loc[[prod],['price ' + str(count)]] = all_days_prices.loc[all_days_prices['PRODUCT'] == prod]['PRICE'].iloc[count]                      
        except:
            #If we exceed each product´s index limit we´ll break and pass to the next product
            break
```

```python
#Display all columns
#pd.set_option('display.max_columns', None)
```

```python
uniques_2
```

```python
len(uniques_2)
```

```python
#Create new df using only the products that have more than 9 price points in the historical series
uniques_2_more_obs = uniques_2[uniques_2['count']>9]
uniques_2_more_obs
```

```python
len(uniques_2_more_obs)
```

```python
#Create final df that will be used to plot the graphics
uniques_2_more_obs_final = uniques_2_more_obs.drop('count', axis=1)
```

```python
import numpy as np
```

```python
#Create function that will remove '€' symbol, select only last price shown (the highest price) and tranform into float type
def format_prices(string):
    try:
        return float(string.split('€')[-1])
    except:
        return np.nan
```

```python
#Apply format_prices func to all 'uniques_2_more_obs_final' df 
for c in uniques_2_more_obs_final.columns:
    uniques_2_more_obs_final[c] = uniques_2_more_obs_final.apply(lambda x: format_prices(x[c]), axis=1)    
```
